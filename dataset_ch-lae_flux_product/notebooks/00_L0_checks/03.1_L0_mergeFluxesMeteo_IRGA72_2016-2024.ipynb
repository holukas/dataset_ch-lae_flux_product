# Merge L0 fluxes and meteo data (2016-2024)

```python
import diive as dv
import pandas as pd
from datetime import datetime
from diive.core.io.files import load_parquet, save_parquet
```

# Load L0 fluxes

```python
SOURCEFILE = r"01.2_OPENLAG_EDDYPRO_FLUXNET_OUTPUT_IRGA72_2016-2024.parquet"
fluxes = load_parquet(filepath=SOURCEFILE)
keeplocs = (fluxes.index.year >= 2017) & (fluxes.index.year <= 2024)
fluxes = fluxes[keeplocs]
# keepcols = ['FC']
# fluxes = fluxes[keepcols]
fluxes
```

# Load meteo

```python
SOURCEFILE = r"../10_METEO/12.5_METEO7_GAPFILLED_2004-2024.parquet"
meteo = load_parquet(filepath=SOURCEFILE)
keeplocs = (meteo.index.year >= 2017) & (meteo.index.year <= 2024)
meteo = meteo[keeplocs]
# keepcols = ['TA_T1_47_1_gfXG', 'VPD_T1_47_1_gfXG', 'SW_IN_T1_47_1_gfXG']
# meteo = meteo[keepcols]
meteo
```

# Merge data

```python
merged = pd.concat([fluxes, meteo], axis=1)
merged
```

# Save to file

```python
OUTNAME = "03.2_FC_OPENLAG_EDDYPRO_FLUXNET_OUTPUT_IRGA72_2016-2024"
OUTPATH = r""
filepath = save_parquet(filename=OUTNAME, data=merged, outpath=OUTPATH)
# merged.to_csv(Path(OUTPATH) / f"{OUTNAME}.csv")
```
