# Generate link to notebooks for Jupyter book
- https://jupyterbook.org/en/stable/structure/configure.html#specify-alternate-titles

```python
# This link worked on the intro.md page:
# [test](https://holukas.github.io/dataset_ch-cha_flux_product/notebooks/20_MANAGEMENT/22.0_ConvertMgmtToTimeseries.html)

import os
import pathlib
from pathlib import Path
from diive.core.io.filereader import search_files

LOCATION = "notebooks"
notebooks = search_files(searchdirs=f"./", pattern="*.ipynb")
notebooks = [n for n in notebooks if ".ipynb_checkpoints" not in str(n)]
notebooks = [n for n in notebooks if "x-" not in str(n)]
notebooks = [n for n in notebooks if "tests" not in str(n)]
notebooks = [n for n in notebooks if not str(n.stem).startswith("_")]

print(notebooks)

# Working link in book:
# outpath = "dataset_ch-cha_flux_product/notebooks/"
outpath = "https://holukas.github.io/dataset_ch-lae_flux_product/notebooks/"
# https://holukas.github.io/holukas.github.io/dataset_ch-cha_flux_product/notebooks/20_MANAGEMENT/22.0_ConvertMgmtToTimeseries.html
# outpath = "https://holukas.github.io/dataset_ch-cha_flux_product/notebooks/20_MANAGEMENT/22.0_ConvertMgmtToTimeseries.html"

chapter = ""

with open("README.md", "w") as f:    
    f.write("# Notebook overview\n")
        
    for nb in notebooks:           
        
        rel_filepath = nb.parent
        this_chapter = rel_filepath        
        
        if this_chapter != chapter:
            chapter = this_chapter
            f.write(f"\n\n## {chapter}\n")
            print(chapter)
            
        file_extension = nb.suffix      
        filename = nb.name
        filename_noext = nb.stem               
        filelink = outpath / rel_filepath / f"{filename_noext}.html"
        # filelink = Path("notebooks") / rel_filepath / f"{filename_noext}.html"
        # filelink = f"{filename_noext}.html"
        
        # Browser for html uses forward slash, also on Windows
        # However, this also changes https:// to https:/, which is not desired, so fix that part later
        filelink = filelink.as_posix()  
        # filelink = Path(outpath) / filelink
        
        link = f"{filelink}"
        link = link.replace("https:/", "https://")
        linestr = f"- [{filename_noext}]({link})\n"
        # linestr = f"{filelink}\n"
        f.write(linestr)
        print(linestr)

print(f"\nLinks generated and saved to README")
```
